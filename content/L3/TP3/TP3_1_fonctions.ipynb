## Qu'est-ce qu'une fonction ?

Une **fonction** est un bout de code Python réutilisable qui :

- prend des **paramètres** en entrée (optionnel)
- effectue un **traitement**
- retourne un **résultat** (optionnel)

## Syntaxe de base

```python
def nom_fonction(parametre1, parametre2):
    # instructions dans la fonction
    return resultat
```

- Une fonction est déclarée par le mot clé **`def`**  et cette première ligne se termine par **`:`**.
- Les instructions exécutées à l'intérieur de la fonction sont dans un bloc d'instructions indenté, c'est-à-dire décalé vers la droite de 4 espaces (pour les boucles avec `for`, les tests avec `if` et les fichiers avec `with`).
- Pour qu'une fonction renvoie quelque chose, on utilise le mot clé **`return`**. Ce mot clé n'est pas obligatoire, car une fonction ne renvoie pas toujours quelque chose.

## Premiers exemples

Exécutez la cellule ci-dessous pour définir la fonction, puis la fonction pour l'exécuter.

````python
def dire_bonjour():
    print("Bonjour, bienvenue en Python !")
````

````python
dire_bonjour()
````

Remarques :

- La fonction `dire_bonjour()` ne prend pas d'argument, c'est-à-dire qu'on ne met rien dans les parenthèses quand on utilise cette fonction.
- Cette fonction ne renvoie rien, car elle ne contient pas le mot-clé `return`. Par contre, elle fait bien quelque chose puisqu'elle affiche un message.

Voici maintenant une fonction qui prend un argument.

Exécutez la cellule suivante pour créer cette fonction.

````python
def dire_bonjour_2(prenom):
    print(f"Bonjour {prenom}, bienvenue en Python !")
````

Appelez maintenant cette fonction en mettant votre prénom dans les parenthèses de la fonction :

````python
dire_bonjour_2("votre prénom")
````

Enfin, voici une fonction qui prend en argument deux nombres et renvoie quelque chose :

````python
def surface_rectangle(longueur, largeur):
    return longueur * largeur
````

On appelle cette fonction en indiquant deux nombres séparés par une virgule dans les parenthèses.

Par exemple :

````python
surface_rectangle(3, 4)
````

On peut utiliser des variables comme arguments de la fonction.

Par exemple, si on définit :

````python
longueur_mesuree = 3.2
largeur_mesuree = 4.7
````

On appelle la fonction `surface_rectangle()` avec ces variables :

````python
surface_rectangle(longueur_mesuree, largeur_mesuree)
````

Ici, ce que renvoie la fonction est directement affiché dans le *notebook*. Mais on peut stocker ce que renvoie la fonction dans une variable :

````python
surface_studio = surface_rectangle(4, 3.5)
````

Et ensuite afficher le contenu de la variable qui a stoché le résultat de la fonction :

````python
print(surface_studio)
````

On peut utiliser une fonction qui renvoie quelque chose directement avec une écriture formatée (*f-string*) :

````python
print(f"La surface de l'appartement est de {surface_rectangle(5.3, 7.3)} m²")
````

````python
longueur_mesuree = 5.25
largeur_mesuree = 6.73
print(f"La surface de l'appartement est de {surface_rectangle(longueur_mesuree, largeur_mesuree)} m²")
````

Ou avec moins de décimales :

````python
longueur_mesuree = 5.25
largeur_mesuree = 6.73
print(f"La surface de l'appartement est de {surface_rectangle(longueur_mesuree, largeur_mesuree):.1f} m²")
````

### Exercice 4

Créez une fonction `convertir_F_vers_C()` qui prend en argument une température en degré Fahrenheit et renvoie la température correspondante en degré Celsius.

On rappelle la formule qui permet de passer d'une température en degré Fahrenheit ($T_F$) à une température en degré Celsius ($T_C$) (voir aussi le *notebook* `TP1_1_variables_operations.ipynb`) :

$$T_C = (T_F - 32) \times \frac{5}{9}$$

Utilisez ensuite cette fonction dans une boucle `for` pour calculer les températures en degré Celsius qui correspondent à 98.6 °F et 212 °F et afficher le message :

```none
XX °F correspond à YY °C
```

avec `XX` la température en degré Fahrenheit (°F) et `YY` la température correspondante en degré Celsius (°C).

````python
def convertir_F_vers_C(temperature_en_F):
    temperature_en_C = (temperature_en_F - 32) * 5/9
    return temperature_en_C
````

````python
for temperature_en_F in [98.6, 212]:
    temperature_en_C = convertir_F_vers_C(temperature_en_F)
    print(f"{temperature_en_F} °F correspond à {temperature_en_C} °C")
````
