## Rappel sur la cinétique enzymatique de type Michaelis-Menten

Une enzyme michaelienne E transforme un substrat S en produit P selon l'équation :

$$
E + S \Longleftrightarrow ES \rightarrow E + P
$$

On suppose ici que [S] >> [E].

La vitesse de la réaction enzymatique lors de la phase stationnaire, $v$, dépend de la concentration en substrat $[S]$ selon la loi de Michaelis-Menten :

$$
v = \frac{V_{max} \times [S]}{K_m + [S]}
$$

avec :

- $V_{max}$ : vitesse initiale quand l’enzyme est saturée en substrat.
- $K_m$ : constante de Michaelis, concentration initiale en substrat $[S]_0$ pour laquelle la vitesse vaut $V_{max}/2$.

Par définition, on peut écrire que de la vitesse de réaction enzymatique est la variation de la concentration en produit [P] en fonction du temps $t$ :

$$
v = \frac{ d[P] }{ dt }
$$

On peut aussi écrire une formule équivalente pour le substrat :

$$
v = - \frac{ d[S] }{ dt }
$$

Le signe **-** rappelle ici que la concentration en substrat diminue au cours du temps.

La variation de le concentration du produit $d[P]$ est équivalente à la différence entre la concentration du produit à temps $t + dt$, notée $[P]_{t + dt}$ et la concentration du produit à temps $t$, notée $[P]_{t}$.

La vitesse vaut donc :

$$
v = \frac{ [P]_{t + dt} - [P]_{ dt } }{ dt } = - \frac{ [S]_{t + dt} - [S]_{ dt } }{ dt }
$$

La concentration en substrat [S] et en produit [P] peut alors s'écrire au temps $t + dt$ :

$$
[P]_{t + dt} = [P]_{t} + v \times ((t + dt) - t) = [P]_{t} + v \times dt
$$

et

$$
[S]_{t + dt} = [S]_{t} - v \times ((t + dt) - t) = [S]_{t} - v \times dt
$$

Numériquement, on peut représenter l'évolution de la [S] et [P] en fonction du temps en connaissant $V_{max}$ et $K_m$.

Supposons ici, de façon totalement arbitraire que :

- $V_{max} = 3,5$
- $K_m = 2$
- $[S]_0 = 10$
- et bien sur $[P] = 0$

On définit d'abord une fonction `vitesse_michaelis_menten()` qui va calculer la vitesse de la réaction en fonction de [S], $V_{max}$ et $K_m$ :

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)
```

Pourquoi utilise-t-on `S` au lieu de `[S]` pour la variable correspondante à la concentration en produit ?

-> En Python, `[S]` est une liste contenant un élément. Cet élément est `S` qui n'est pas défini.

On définit maintenant une fonction `calcule_cinetique()` qui va calculer les concentrations du substrat et du produit à différents pas de temps. Cette fonction renvoie 3 listes qui correspondent aux différents temps, à la concentration du substrat à chaque temps et à la concentration du produit à chaque temps.

À chaque fois, on calcule la nouvelle concentration du substrat ou du produit à partir de la vitesse de réaction et la dernière concentration en substrat ou en produit.

```python
def calcule_cinetique(S_0, Vmax, Km, pas_temps, nombre_points):
    dt = pas_temps
    # On initalise le premier temps à 0 (le début de la simulation).
    temps = [0]
    # On initialise la première concentration de substrat avec S_0,
    # c'est-à-dire la concentration en substrat au début de l'expérience.
    concentrations_substrat = [S_0]
    # On initialise la première concentration en produit à 0,
    # car il n'y a pas de produit au début de la cinétique.
    concentrations_produit = [0]
    # Pour chaque point, le temps avance de dt.
    # On calcule à chaque fois le nouveau temps, la nouvelle concentration en susbtrat
    # et la nouvelle constration en produit
    for point in range(0, nombre_points):
        temps.append(temps[-1] + dt)
        v = vitesse_michaelis_menten(concentrations_substrat[-1], Vmax, Km)
        concentrations_substrat.append(concentrations_substrat[-1] - v*dt)
        concentrations_produit.append(concentrations_produit[-1] + v*dt)

    return temps, concentrations_substrat, concentrations_produit
```

Testons cette fonction avec un pas de temps de `0.1` et `100` points de temps :

```python
liste_temps, liste_conc_susbtrat, liste_conc_produit = calcule_cinetique(10, 3.5, 2, 0.1, 100)
```

Et affichons graphiquement le résultat : 

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots()
ax.set_xlabel("Temps")
ax.set_ylabel("Concentration")
ax.set_title("Concentration en fonction du temps")
ax.plot(liste_temps, liste_conc_susbtrat, color="green", label="substrat")
ax.plot(liste_temps, liste_conc_produit, color="orange",  label="produit")
ax.legend()
plt.show()
```

Dans la cellule ci-dessous, faites varier la concentration initiale en substrat, $V_{max}$ ou $K_m$ pour comprendre l'influence de ces paramètres sur la cinétique de la réaction.
