
# Activité 2 — Calculer une vitesse enzymatique en Python

On va écrire une fonction qui calcule la vitesse.

## Code Python

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    v = (Vmax * S) / (Km + S)
    return v

# Test
Vmax = 10
Km = 2

S = 1
v = vitesse_michaelis_menten(S, Vmax, Km)
print("Vitesse =", v)
```

## Question

Que vaut la vitesse quand S = Km ?

## Réponse détaillée

Si S = Km :

$$
v = \frac{V_{max} \times K_m}{K_m + K_m}
= \frac{V_{max} \times K_m}{2K_m}
= \frac{V_{max}}{2}
$$

Donc **Km est la concentration pour laquelle la vitesse vaut la moitié de Vmax**.
Biologiquement, cela correspond à une enzyme occupée à **50%** par le substrat.

# Activité 4 — Simuler l’évolution du produit au cours du temps

On veut maintenant simuler une expérience où on mesure la quantité de produit formé au cours du temps.

Principe :
[
\frac{dP}{dt} = v
]

Donc numériquement :

[
P_{t+dt} = P_t + v \times dt
]

## Code Python

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

# Paramètres
Vmax = 10
Km = 2
S = 5      # concentration initiale
P = 0      # produit initial
dt = 0.1   # pas de temps
temps_total = 5

temps = 0

liste_temps = []
liste_P = []

while temps <= temps_total:
    v = vitesse_michaelis_menten(S, Vmax, Km)
    
    P = P + v * dt
    S = S - v * dt   # le substrat diminue
    
    liste_temps.append(temps)
    liste_P.append(P)
    
    temps = temps + dt

print(liste_P)
```

# Activité 5 — Interprétation biologique

## Question

Pourquoi la vitesse diminue au cours du temps dans la simulation ?

## Réponse détaillée

Dans la simulation, on a :

$$
S = S - v \times dt
$$

Donc la concentration en substrat diminue au cours du temps car il est transformé en produit.

Or la vitesse dépend de [S] :

$$
v = \frac{V_{max} [S]}{K_m + [S]}
$$

Donc quand [S] diminue :

* le numérateur diminue
* la vitesse diminue

**Interprétation biologique :**

* Au début : beaucoup de substrat → réaction rapide
* À la fin : presque plus de substrat → réaction lente
* C’est exactement ce qu’on observe dans une vraie expérience enzymatique

# Activité 7 — Tracer une courbe de Michaelis-Menten

En biologie, on trace la courbe :

[
v = f([S])
]

Elle a une forme **hyperbolique** :

* Au début : la vitesse augmente vite
* Puis : la vitesse atteint un plateau = **Vmax**
* Cela correspond à la **saturation de l’enzyme**

## Objectif informatique

On veut calculer beaucoup de vitesses pour beaucoup de valeurs de [S].

## Code Python

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

Vmax = 10
Km = 2

S = 0
pas = 0.5

liste_S = []
liste_v = []

while S <= 20:
    v = vitesse_michaelis_menten(S, Vmax, Km)
    
    liste_S.append(S)
    liste_v.append(v)
    
    S = S + pas

print(liste_S)
print(liste_v)
```

Ici on a simulé une expérience où on teste **plein de concentrations en substrat**.

---

## Question

Pourquoi la courbe atteint-elle un plateau ?

## Réponse détaillée

Quand [S] devient très grand :

[
v = \frac{V_{max} [S]}{K_m + [S]}
]

Si [S] >> Km, alors Km devient négligeable :

[
v \approx \frac{V_{max} [S]}{[S]} = V_{max}
]

Donc la vitesse ne dépend plus de [S].

**Interprétation biologique :**

* Toutes les enzymes sont occupées par un substrat
* L’enzyme travaille à sa vitesse maximale
* Ajouter du substrat ne change plus la vitesse

C’est la **saturation enzymatique**.

# Activité 8 — Déterminer Km et Vmax à partir de données expérimentales

En vrai TP, on ne connaît pas Km et Vmax.
On doit les **estimer** à partir des mesures expérimentales.

On va tester plusieurs valeurs et voir lesquelles donnent les résultats les plus proches des données expérimentales.

## Données expérimentales (inventées)

```python
S_exp = [0.5, 1, 2, 4, 6, 8, 10]
v_exp = [2.0, 3.7, 5.2, 6.5, 7.3, 8.1, 8.5]
```

---

## Idée mathématique

On va calculer l’erreur :

[
erreur = \sum (v_{exp} - v_{theo})^2
]

Plus l’erreur est petite → meilleur modèle.

---

## Code Python

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

def erreur_modele(S_exp, v_exp, Vmax, Km):
    erreur = 0
    
    i = 0
    while i < len(S_exp):
        v_theo = vitesse_michaelis_menten(S_exp[i], Vmax, Km)
        diff = v_exp[i] - v_theo
        erreur = erreur + diff * diff
        i = i + 1
    
    return erreur

S_exp = [0.5, 1, 2, 4, 6, 8, 10]
v_exp = [2.0, 3.7, 5.2, 6.5, 7.3, 8.1, 8.5]

print(erreur_modele(S_exp, v_exp, 10, 2))
```

On élève au carré pour :

2. Donner plus d’importance aux grosses erreurs
3. Méthode mathématique classique : **méthode des moindres carrés**

# Activité 9 — Trouver automatiquement Km et Vmax

On va tester plein de valeurs et trouver la meilleure.

## Code Python

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

def erreur_modele(S_exp, v_exp, Vmax, Km):
    erreur = 0
    i = 0
    while i < len(S_exp):
        v_theo = vitesse_michaelis_menten(S_exp[i], Vmax, Km)
        diff = v_exp[i] - v_theo
        erreur = erreur + diff * diff
        i = i + 1
    return erreur

S_exp = [0.5, 1, 2, 4, 6, 8, 10]
v_exp = [2.0, 3.7, 5.2, 6.5, 7.3, 8.1, 8.5]

meilleure_erreur = 100000
meilleur_Vmax = 0
meilleur_Km = 0

Vmax = 5
while Vmax <= 15:
    Km = 0.5
    while Km <= 10:
        
        err = erreur_modele(S_exp, v_exp, Vmax, Km)
        
        if err < meilleure_erreur:
            meilleure_erreur = err
            meilleur_Vmax = Vmax
            meilleur_Km = Km
        
        Km = Km + 0.5
    Vmax = Vmax + 0.5

print("Vmax =", meilleur_Vmax)
print("Km =", meilleur_Km)
```

## Réponse détaillée

Ce programme simule ce que fait un biologiste quand il cherche Km et Vmax :

* Il teste différentes valeurs de Km et Vmax
* Il compare aux données expérimentales
* Il garde les valeurs qui expliquent le mieux l’expérience

# Activité 10 — Inhibition enzymatique

En biologie, il existe des inhibiteurs :

* compétitif
* non compétitif
* incompétitif

### Inhibition compétitive

Elle augmente **Km** mais ne change pas **Vmax**.

Nouvelle formule :

[
v = \frac{V_{max} [S]}{K_m(1 + [I]/K_i) + [S]}
]

où :

* I = concentration inhibiteur
* Ki = constante d’inhibition

---

## Code Python

```python
def vitesse_inhibition_competitive(S, Vmax, Km, I, Ki):
    Km_apparent = Km * (1 + I / Ki)
    v = (Vmax * S) / (Km_apparent + S)
    return v

Vmax = 10
Km = 2
I = 1
Ki = 1

S = 0.5

while S <= 10:
    v = vitesse_inhibition_competitive(S, Vmax, Km, I, Ki)
    print(S, v)
    S = S + 0.5
```

---

## Question

Pourquoi Km augmente avec un inhibiteur compétitif ?

## Réponse détaillée

Un inhibiteur compétitif :

* ressemble au substrat
* se fixe sur le site actif
* empêche le substrat de se fixer

Donc il faut **plus de substrat** pour atteindre la même vitesse.

Donc :

* L’affinité apparente enzyme-substrat diminue
* Donc **Km augmente**
* Mais si on met énormément de substrat → on peut déplacer l’inhibiteur → donc **Vmax ne change pas**

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

def erreur_modele(S_exp, v_exp, Vmax, Km):
    erreur = 0
    i = 0
    while i < len(S_exp):
        v_theo = vitesse_michaelis_menten(S_exp[i], Vmax, Km)
        diff = v_exp[i] - v_theo
        erreur = erreur + diff**2
        i = i + 1
    return erreur

S_exp = [0.5, 1, 2, 3, 5, 7, 10]
v_exp = [1.5, 2.8, 4.5, 5.5, 6.8, 7.5, 8.2]

meilleure_erreur = 1000
meilleur_Vmax = 0
meilleur_Km = 0

Vmax = 5
while Vmax <= 12:
    Km = 0.5
    while Km <= 5:
        
        err = erreur_modele(S_exp, v_exp, Vmax, Km)
        
        if err < meilleure_erreur:
            meilleure_erreur = err
            meilleur_Vmax = Vmax
            meilleur_Km = Km
        
        Km = Km + 0.5
    Vmax = Vmax + 0.5

print("Vmax =", meilleur_Vmax)
print("Km =", meilleur_Km)
```

---

## Interprétation biologique (réponse attendue)

* **Vmax** = vitesse maximale quand l’enzyme est saturée
* **Km** = concentration en substrat pour laquelle v = Vmax/2
* Si Km petit → enzyme a forte affinité pour le substrat
* Si Km grand → enzyme a faible affinité

Simuler l’évolution de S et P au cours du temps.

## Correction

```python
def vitesse_michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

S = 8
P = 0
dt = 0.1
temps = 0
temps_total = 6

Vmax = 9
Km = 2

while temps <= temps_total:
    v = vitesse_michaelis_menten(S, Vmax, Km)
    
    P = P + v * dt
    S = S - v * dt
    
    print(temps, S, P)
    
    temps = temps + dt
```

## Question

À quoi ressemble la courbe P = f(t) ?

## Réponse détaillée

La courbe :

* augmente rapidement au début
* puis ralentit
* puis atteint un plateau

Pourquoi ?

* Au début : beaucoup de substrat → réaction rapide
* Ensuite : substrat diminue → réaction ralentit
* À la fin : plus de substrat → plus de réaction

C’est exactement la forme observée dans une vraie réaction enzymatique.

# Partie 4 — Inhibiteur compétitif

On travaille maintenant avec un inhibiteur.

Formule :

[
v = \frac{V_{max} S}{K_m(1 + I/K_i) + S}
]

## Énoncé

Simuler la réaction avec :

* I = 2
* Ki = 1
  Comparer avec la réaction sans inhibiteur.

---

## Correction

```python
def vitesse_inhibition(S, Vmax, Km, I, Ki):
    Km_app = Km * (1 + I / Ki)
    return (Vmax * S) / (Km_app + S)

S = 8
P = 0
dt = 0.1
temps = 0
temps_total = 6

Vmax = 9
Km = 2
I = 2
Ki = 1

while temps <= temps_total:
    v = vitesse_inhibition(S, Vmax, Km, I, Ki)
    
    P = P + v * dt
    S = S - v * dt
    
    print(temps, S, P)
    
    temps = temps + dt
```

---

## Interprétation biologique (réponse attendue)

Avec inhibiteur compétitif :

* La réaction est plus lente
* Il faut plus de substrat pour atteindre la même vitesse
* Donc **Km augmente**
* **Vmax ne change pas**
