# Détermination de Vmax et Km

La plupart du temps, on ne connait pas les valeurs de $V_{max}$ et $K_m$. On peut cependant les déterminer expérimentalement, notamment en mesurant la vitesse initiale de cinétique pour différentes concentrations initiales de substrat. Cette méthode est très efficace. Nous allons simuler cette approche numériquement avec Python.

En reprenant la loi de Michaelis-Menten, la vitesse initiale de la réaction enzymatique, $v_i$, s'exprime en fonction de la concentration initiale en substrat $[S]_0$ selon :

$$
v_i = \frac{V_{max} \times [S]_0}{K_m + [S]_0}
$$

Interprétation biologique :

- Si $[S]_0 << K_m$, la vitesse initiale est proportionnelle à [S].
- Si $[S]_0 >> K_m$, la vitesse initiale ≈ $V_{max}$ (saturation enzymatique).

Expérimentalement, on mesure la vitesse initiale $v_i$ pour plusieurs concentrations initiales en substrat $[S]_0$. Pour cela, on réalise plusieurs cinétiques avec des concentrations initiales en substrat différentes, et pendant lesquelles on mesure la concentration du produit en fonction du temps.
