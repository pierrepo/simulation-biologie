## La fonction `print()`

Comme son nom l'indique, cette fonction est utilisée pour afficher quelque chose.

On peut afficher des chaînes de caractères :

```python
print("Salut Python")
```

Des nombres entiers :

```python
print(42)
```

Des nombres réels :

```python
nombre = 3.14
```

```python
print(nombre)
```

On peut aussi afficher plusieurs variables. Soit séparémment :

```python
seq = "ATCTGA" 
nombre_bases = 6
```

```python
print(seq)
print(nombre_bases)
```

Soit ensemble, en séparant le nom de chaque variable par une virgule :

```python
print(seq, nombre_bases)
```

On peut même alterner des chaînes de caractères avec des noms de variables :

```python
print("La séquence", seq, "contient", nombre_bases, "bases.")
```

## Écriture formatée (*f-strings*)

L'écriture formatée apporte un contrôle plus fin sur la manière d'afficher les variables.

On utilise une grande chaîne de caractères avec à l'intérieur des variables entre accolades :

```python
print(f"La séquence {seq} contient {nombre_bases} bases.")
```

:::{admonition} Attention
:class: warning

- au caractère **`f`** (comme *f-strings*) devant la chaîne de caractères (juste avant le premier guillemet)
- aux noms des variables entre **`{ }`**

::::

### Contrôler le nombre de chiffres affichés après la virgule

```python
valeur_experimentale = 34.7879783705980890
print(valeur_experimentale)
```

Vous l'avez sans doute appris en TP, il n'est pas pertinent d'afficher une valeur numérique avec tous les chiffres après la virgule.

L'écriture formatée permet de limiter le nombre de chiffres après la virgule.

Avec un seul chiffre après la virgule :

```python
print(f"{valeur_experimentale:.1f}")
```

📝 Remarques : 

- Le symbole `:` sépare le nom de la variable (`valeur_experimentale`) de l'opérateur de formatage (`.1f`).
- L'opérateur de formatage `.1f` signifie qu'on souhaite afficher un nombre réel (*float* -> `f`) avec un chiffre après la virgule (`.1`)
- C'est bien un arrondi qui est réalisé ici.

Avec deux chiffres après la virgule :

```python
print(f"{valeur_experimentale:.2f}")
```

Avec trois chiffres après la virgule :

```python
print(f"{valeur_experimentale:.3f}")
```
